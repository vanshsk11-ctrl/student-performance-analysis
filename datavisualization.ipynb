```python
#Import needed libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv("StudentPerformance.csv")
```

```python
df.columns = [
    "hours_studied",
    "previous_scores",
    "extracurricular",
    "sleep_hours",
    "practice_papers",
    "performance_index"
]


df["extracurricular"].unique()
df["extracurricular"].value_counts()

df["extracurricular"] = (
    df["extracurricular"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": 1, "no": 0})
)


print(df["extracurricular"].value_counts())
```

```python

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['performance_index'], bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Performance Index')
plt.ylabel('Frequency')
plt.title('Distribution of Student Performance Index')
plt.show()
```

```python
#Scatterplot to show the correlation of hours studied and Student performance
plt.figure(figsize=(10, 6))
plt.scatter(df['hours_studied'], df['performance_index'], alpha=0.5)
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.title('Study Hours vs Performance Index')
plt.show()
```

```python
#Visualizes how each variable correlates to eachother
numeric_cols = df[['hours_studied', 'previous_scores', 'extracurricular',
                   'sleep_hours', 'practice_papers', 'performance_index']]

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', center=0,
            fmt='.2f', square=True, linewidths=1)
plt.title('Correlation Matrix of Student Performance Factors')
plt.tight_layout()
plt.show()
```

```python
#Boxplot to compare how extracurriculars affect performance
plt.figure(figsize=(8, 6))
sns.boxplot(x='extracurricular', y='performance_index', data=df)
plt.xlabel('Extracurricular Activities (0=No, 1=Yes)')
plt.ylabel('Performance Index')
plt.title('Impact of Extracurricular Activities on Performance')
plt.show()
```

```python
#Scatterplots for the relationship of each numerical variable and performance
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Hours studied
axes[0, 0].scatter(df['hours_studied'], df['performance_index'], alpha=0.5)
axes[0, 0].set_xlabel('Hours Studied')
axes[0, 0].set_ylabel('Performance Index')
axes[0, 0].set_title('Hours Studied vs Performance')

# Sleep hours
axes[0, 1].scatter(df['sleep_hours'], df['performance_index'], alpha=0.5, color='green')
axes[0, 1].set_xlabel('Sleep Hours')
axes[0, 1].set_ylabel('Performance Index')
axes[0, 1].set_title('Sleep Hours vs Performance')

# Previous scores
axes[1, 0].scatter(df['previous_scores'], df['performance_index'], alpha=0.5, color='orange')
axes[1, 0].set_xlabel('Previous Scores')
axes[1, 0].set_ylabel('Performance Index')
axes[1, 0].set_title('Previous Scores vs Performance')

# Practice papers
axes[1, 1].scatter(df['practice_papers'], df['performance_index'], alpha=0.5, color='red')
axes[1, 1].set_xlabel('Practice Papers')
axes[1, 1].set_ylabel('Performance Index')
axes[1, 1].set_title('Practice Papers vs Performance')

plt.tight_layout()
plt.show()
```
